```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class GradientBoosting:
  def __init__(self, alpha = 0.1, max_depth = 4, n_trees = 100):
    self.alpha = alpha
    self.max_depth = max_depth
    self.n_tree = n_trees

    self.trees = []
    self.initial_value = None

  def fit(self, X, y):
    self.initial_value = np.mean(y)
    pred = np.full(len(y), self.initial_value)
    for _ in range(self.n_tree):
      residual = y - pred
      tree = DecisionTreeRegressor(max_depth=self.max_depth)
      tree.fit(X,residual)
      update = tree.predict(X)

      pred += self.alpha*update
      self.trees.append(tree)

  def predict(self, X):
    pred = np.full(X.shape[0], self.initial_value)
    for tree in self.trees:
      update = tree.predict(X)
      pred += self.alpha * update
    return pred


```
