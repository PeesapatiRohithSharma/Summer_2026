```python
ham_count = 2591
spam_count = 409
tot_count = ham_count + spam_count
```

```python
p_spam = (spam_count+1)/(tot_count+2)
p_ham = (ham_count+1)/(tot_count+2)
```

```python
import math
class Naive_Bayes_Classifier:
  def __init__(self, vocab):
    self.vocab = vocab

  def create_email(self,email):
    self.email = re.findall(r'[a-z0-9]+|[!?£$@#%&]', email.lower())
  def calc_probabilities(self,X,y):
    self.p1 = p_spam
    self.p2 = 1
    self.p3 = 1
    sum_1 = 0
    sum_2 = 0
    for word in self.email:
      counter_1 = 0
      counter_0 = 0
      for i in range(len(X)):
        tokens_i = set(re.findall(r'[a-z0-9]+|[!?£$@#%&]', X[i].lower()))
        if word in tokens_i:
          if y[i] == 1:
              counter_1 += 1
          else:
              counter_0 += 1
      p2 = (counter_1+1)/(spam_count+2)
      p3 = (counter_0+1)/(ham_count+2)
      self.p2 *= p2
      self.p3*= p3

    return self.p1, self.p2, self.p3

  def predict(self):
    prob_spam = self.p2*(self.p1)/(self.p2*(self.p1)+self.p3*(1-self.p1))
    prob_ham = self.p3*(1-self.p1)/(self.p2*(self.p1)+self.p3*(1-self.p1))
    if prob_spam > prob_ham:
      return f"Its a spam email {prob_spam}"

    if prob_spam <= prob_ham:
      return f"Its not a spam {prob_ham}"
```

```python
spam = Naive_Bayes_Classifier(vocab)
email = "I am on my way."
spam.create_email(email)
spam.calc_probabilities(X,y)
spam.predict()
```
